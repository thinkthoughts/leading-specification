## 1. Locate the repository

```python
from pathlib import Path
from dataclasses import dataclass, asdict
from zipfile import ZipFile, ZIP_DEFLATED
import json
import numpy as np
import matplotlib.pyplot as plt

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

FIGURES = ROOT / "figures"
RESULTS = ROOT / "results"
DATA = ROOT / "data"

for path in (FIGURES, RESULTS, DATA):
    path.mkdir(parents=True, exist_ok=True)

def savefig(name):
    path = FIGURES / name
    plt.savefig(path, dpi=240, bbox_inches="tight")
    print("saved:", path)
    return path

print("ROOT:", ROOT)
```

## 2. Define the engineering context

\[
\pi_t \rightarrow T_t \rightarrow d_t \rightarrow I_t
\]

- \(\pi_t\): deployment policy
- \(T_t\): deployment telemetry
- \(d_t\): benefit-distribution state
- \(I_t\): sustainable-development indicators

```python
@dataclass(frozen=True)
class EngineeringContext:
    repository: str
    engineering_statement: str
    constraint: str
    connected_lanes: tuple[str, ...]
    engineering_objects: tuple[str, ...]
    state_variables: tuple[str, ...]
    observations: tuple[str, ...]
    construction_sequence: tuple[str, ...]
    trailing_indicators: tuple[str, ...]
    next_notebook: str
    source: str = ""
    notes: str = ""

    def validate(self):
        for name in ("repository", "engineering_statement", "constraint", "next_notebook"):
            if not getattr(self, name).strip():
                raise ValueError(f"{name} must be non-empty")
        for name in ("connected_lanes", "engineering_objects", "state_variables", "observations", "construction_sequence"):
            values = getattr(self, name)
            if not values or any(not str(v).strip() for v in values):
                raise ValueError(f"{name} must contain non-empty items")

    def save(self, path):
        self.validate()
        path.parent.mkdir(parents=True, exist_ok=True)
        path.write_text(json.dumps(asdict(self), indent=2), encoding="utf-8")
        return path

CONTEXT = EngineeringContext(
    repository="benefits-distribution",
    engineering_statement="Benefit Distribution Specifies Sustainable Development",
    constraint="deployment constraints",
    connected_lanes=("deployment policy","deployment telemetry","benefit distribution","development indicators"),
    engineering_objects=("policy","telemetry","distribution state","indicator"),
    state_variables=("pi_t","T_t","d_t","I_t"),
    observations=("deployment telemetry",),
    construction_sequence=(
        "00 engineering context",
        "07 telemetry",
        "13 state estimation",
        "17 distribution",
        "23 dynamics",
        "29 optimization",
        "37 controllers",
        "43 benchmarks",
    ),
    trailing_indicators=("sustainable-development indicators",),
    next_notebook="07_telemetry.ipynb",
    source="Google DeepMind: The Case for Globally Beneficial Technology",
    notes="Standalone Notebook 00 for the benefits-distribution repository.",
)

CONTEXT.validate()
CONTEXT
```

## 3. Save the context manifest

```python
MANIFEST = CONTEXT.save(RESULTS / "00_engineering_context.json")
print("saved:", MANIFEST)
```

## 4. Engineering grammar

Notebook 00 begins with a shared engineering grammar:

\[
\text{Constraint}
\rightarrow
\text{Connected lanes}
\rightarrow
\text{Engineering objects}
\rightarrow
\text{Observable states}
\rightarrow
\text{Indicators}
\]

The deployment-specific lane is one application of this reusable structure.

```python
# Figure 1 — Engineering grammar

grammar = [
    "Constraint",
    "Connected lanes",
    "Engineering objects",
    "Observable states",
    "Indicators",
]

plt.figure(figsize=(8, 8.5))
ax = plt.gca()
ax.axis("off")

ys = np.linspace(0.82, 0.20, len(grammar))
x = 0.5

for i, (label, y) in enumerate(zip(grammar, ys)):
    width = 0.50 if len(label) < 19 else 0.60
    ax.add_patch(
        plt.Rectangle(
            (x - width / 2, y - 0.045),
            width,
            0.09,
            fill=False,
            linewidth=1.8,
        )
    )
    ax.text(
        x,
        y,
        label,
        ha="center",
        va="center",
        fontsize=13,
        fontweight="bold" if i in (0, 1, 2) else None,
    )

    if i < len(grammar) - 1:
        ax.annotate(
            "",
            xy=(x, ys[i + 1] + 0.055),
            xytext=(x, y - 0.055),
            arrowprops=dict(
                arrowstyle="->",
                linewidth=1.8,
            ),
        )

ax.text(
    0.5,
    0.94,
    "Engineering Grammar",
    ha="center",
    fontsize=18,
    fontweight="bold",
)

ax.text(
    0.5,
    0.08,
    "Leading constraints specify connected lanes. "
    "Connected lanes specify engineering objects. "
    "Indicators follow observable states.",
    ha="center",
    fontsize=10.5,
)

savefig("00_engineering_grammar.png")
plt.show()
```

## 5. Connected deployment lanes

\[
\pi_t \rightarrow T_t \rightarrow d_t \rightarrow I_t
\]

Leading constraints specify connected lanes. Connected lanes specify benefit distribution. Sustainable development follows benefit distribution.

```python
labels = [
    (r"$\pi_t$", "deployment policy"),
    (r"$T_t$", "deployment telemetry"),
    (r"$d_t$", "benefit distribution"),
    (r"$I_t$", "development indicators"),
]

plt.figure(figsize=(12, 4.8))
ax = plt.gca()
ax.axis("off")
xs = np.linspace(0.12, 0.88, len(labels))
y = 0.54

for i, (x, (symbol, label)) in enumerate(zip(xs, labels)):
    ax.add_patch(plt.Rectangle((x-0.075, y-0.09), 0.15, 0.18, fill=False, linewidth=1.8))
    ax.text(x, y+0.025, symbol, ha="center", va="center", fontsize=20)
    ax.text(x, y-0.055, label, ha="center", va="center", fontsize=9)
    if i < len(labels)-1:
        ax.annotate("", xy=(xs[i+1]-0.09, y), xytext=(x+0.09, y),
                    arrowprops=dict(arrowstyle="->", linewidth=1.8))

ax.text(0.5, 0.90, "Connected Deployment Lanes", ha="center", fontsize=18, fontweight="bold")
ax.text(0.5, 0.16,
        "Leading constraints specify connected lanes. Connected lanes specify benefit distribution. Sustainable development follows benefit distribution.",
        ha="center", fontsize=11)

savefig("00_connected_deployment_lanes.png")
plt.show()
```

## 6. Construction sequence

```python
sequence = list(CONTEXT.construction_sequence)

plt.figure(figsize=(13, 5))
ax = plt.gca()
ax.axis("off")
xs = np.linspace(0.07, 0.93, len(sequence))
y = 0.53

for i, (x, label) in enumerate(zip(xs, sequence)):
    number, title = label.split(" ", 1)
    ax.add_patch(plt.Rectangle((x-0.055, y-0.10), 0.11, 0.20, fill=False, linewidth=1.8))
    ax.text(x, y+0.045, number, ha="center", va="center", fontsize=13, fontweight="bold")
    ax.text(x, y-0.035, title.replace(" ", "\n", 1), ha="center", va="center", fontsize=8.5)
    if i < len(sequence)-1:
        ax.annotate("", xy=(xs[i+1]-0.065, y), xytext=(x+0.065, y),
                    arrowprops=dict(arrowstyle="->", linewidth=1.7))

ax.text(0.5, 0.91, "Engineering Construction Sequence",
        ha="center", fontsize=18, fontweight="bold")
ax.text(0.5, 0.18,
        "Each notebook develops one engineering object from context through benchmark.",
        ha="center", fontsize=11)

savefig("00_construction_sequence.png")
plt.show()
```
