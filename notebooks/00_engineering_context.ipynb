## 1. Repository specification

Only this cell is repository-specific.

For a future repository, change:

- the repository name and description,
- the engineering statement,
- the repository lane,
- the notebook construction sequence,
- the source and next notebook.

```python
from dataclasses import dataclass, asdict
from pathlib import Path
from zipfile import ZipFile, ZIP_DEFLATED
import json
import numpy as np
import matplotlib.pyplot as plt

@dataclass(frozen=True)
class RepositoryContext:
    repository: str
    repository_description: str
    engineering_statement: str
    constraint: str
    connected_lane: str
    engineering_objects: tuple[str, ...]
    lane_symbols: tuple[str, ...]
    lane_labels: tuple[str, ...]
    engineering_variables: tuple[str, ...]
    observable_states: tuple[str, ...]
    indicators: tuple[str, ...]
    construction_sequence: tuple[str, ...]
    next_notebook: str
    source: str = ""
    notes: str = ""

    def validate(self) -> None:
        for name in (
            "repository",
            "repository_description",
            "engineering_statement",
            "constraint",
            "connected_lane",
            "next_notebook",
        ):
            if not getattr(self, name).strip():
                raise ValueError(f"{name} must be non-empty")

        for name in (
            "engineering_objects",
            "lane_symbols",
            "lane_labels",
            "engineering_variables",
            "observable_states",
            "indicators",
            "construction_sequence",
        ):
            values = getattr(self, name)
            if not values or any(not str(item).strip() for item in values):
                raise ValueError(f"{name} must contain non-empty items")

        if len(self.lane_symbols) != len(self.lane_labels):
            raise ValueError("lane_symbols and lane_labels must have equal length")

CONTEXT = RepositoryContext(
    repository="benefits-distribution",
    repository_description=(
        "This repository specifies benefit distribution as an engineering "
        "variable for sustainable development."
    ),
    engineering_statement="Benefit Distribution Specifies Sustainable Development",
    constraint="deployment constraints",
    connected_lane="policy → telemetry → distribution state → development indicators",
    engineering_objects=(
        "deployment policy",
        "telemetry system",
        "benefit-distribution model",
        "development-indicator system",
    ),
    lane_symbols=(r"$\pi_t$", r"$T_t$", r"$d_t$", r"$I_t$"),
    lane_labels=("policy", "telemetry", "distribution state", "development indicators"),
    engineering_variables=(
        "deployment policy",
        "deployment telemetry",
        "benefit-distribution state",
        "development indicators",
    ),
    observable_states=(
        "observed deployment telemetry",
        "estimated benefit-distribution state",
    ),
    indicators=("sustainable-development indicators",),
    construction_sequence=(
        "00 Engineering context",
        "01 Constraints",
        "07 Connected lanes",
        "11 Engineering variables",
        "13 Telemetry",
        "17 State estimation",
        "19 Distribution",
        "23 Dynamics",
        "29 Optimization",
        "37 Controllers",
        "43 Benchmarks",
    ),
    next_notebook="01_constraints.ipynb",
    source="Google DeepMind: The Case for Globally Beneficial Technology",
    notes="Standalone reusable Notebook 00 specialized for benefits-distribution.",
)

CONTEXT.validate()
CONTEXT
```

## 2. Initialize output paths

This notebook is self-contained and writes only to:

- `figures/`
- `results/`
- `data/`

```python
ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

FIGURES = ROOT / "figures"
RESULTS = ROOT / "results"
DATA = ROOT / "data"

for path in (FIGURES, RESULTS, DATA):
    path.mkdir(parents=True, exist_ok=True)

def savefig(filename: str) -> Path:
    path = FIGURES / filename
    plt.savefig(path, dpi=240, bbox_inches="tight")
    print("saved:", path)
    return path

print("ROOT:", ROOT)
```

## 3. Engineering specification grammar

\[
\text{Constraint}
\rightarrow
\text{Connected lane}
\rightarrow
\text{Engineering object}
\rightarrow
\text{Engineering variable}
\rightarrow
\text{Observable state}
\rightarrow
\text{Indicator}
\]

This grammar is shared by every repository.

```python
grammar = [
    "Constraint",
    "Connected lane",
    "Engineering object",
    "Engineering variable",
    "Observable state",
    "Indicator",
]

plt.figure(figsize=(8, 9.5))
ax = plt.gca()
ax.axis("off")

ys = np.linspace(0.84, 0.18, len(grammar))
x = 0.5

for index, (label, y) in enumerate(zip(grammar, ys)):
    ax.add_patch(
        plt.Rectangle(
            (x - 0.28, y - 0.04),
            0.56,
            0.08,
            fill=False,
            linewidth=1.8,
        )
    )
    ax.text(
        x,
        y,
        label,
        ha="center",
        va="center",
        fontsize=13,
        fontweight="bold" if index < 4 else None,
    )
    if index < len(grammar) - 1:
        ax.annotate(
            "",
            xy=(x, ys[index + 1] + 0.05),
            xytext=(x, y - 0.05),
            arrowprops=dict(arrowstyle="->", linewidth=1.8),
        )

ax.text(0.5, 0.95, "Engineering Specification Grammar", ha="center", fontsize=18, fontweight="bold")
ax.text(
    0.5,
    0.07,
    "Engineering specifies objects before measuring variables, "
    "and measures variables before evaluating indicators.",
    ha="center",
    fontsize=10.5,
)

savefig("00_engineering_specification_grammar.png")
plt.show()
```

## 4. Benefit Distribution as an Engineering Variable

\[
\pi_t
\rightarrow
T_t
\rightarrow
d_t
\rightarrow
I_t
\]

This is the repository-specific figure. Future repositories replace only the symbols, labels, and engineering statement.

```python
symbols = CONTEXT.lane_symbols
labels = CONTEXT.lane_labels

plt.figure(figsize=(12, 4.8))
ax = plt.gca()
ax.axis("off")

xs = np.linspace(0.12, 0.88, len(symbols))
y = 0.54

for index, (x, symbol, label) in enumerate(zip(xs, symbols, labels)):
    ax.add_patch(
        plt.Rectangle(
            (x - 0.075, y - 0.09),
            0.15,
            0.18,
            fill=False,
            linewidth=1.8,
        )
    )
    ax.text(x, y + 0.025, symbol, ha="center", va="center", fontsize=20)
    ax.text(x, y - 0.055, label, ha="center", va="center", fontsize=9)

    if index < len(symbols) - 1:
        ax.annotate(
            "",
            xy=(xs[index + 1] - 0.09, y),
            xytext=(x + 0.09, y),
            arrowprops=dict(arrowstyle="->", linewidth=1.8),
        )

ax.text(
    0.5,
    0.90,
    "Benefit Distribution as an Engineering Variable",
    ha="center",
    fontsize=18,
    fontweight="bold",
)
ax.text(
    0.5,
    0.16,
    "Policy specifies telemetry. Telemetry specifies distribution state. "
    "Distribution state specifies development indicators.",
    ha="center",
    fontsize=10.5,
)

savefig("00_benefit_distribution_engineering_variable.png")
plt.show()
```

## 5. Repository construction sequence

Each notebook specifies one stage of the engineering program.

```python
sequence = CONTEXT.construction_sequence

plt.figure(figsize=(15, 5))
ax = plt.gca()
ax.axis("off")

xs = np.linspace(0.04, 0.96, len(sequence))
y = 0.53

for index, (x, item) in enumerate(zip(xs, sequence)):
    number, title = item.split(" ", 1)

    ax.add_patch(
        plt.Rectangle(
            (x - 0.041, y - 0.10),
            0.082,
            0.20,
            fill=False,
            linewidth=1.6,
        )
    )
    ax.text(x, y + 0.045, number, ha="center", va="center", fontsize=12, fontweight="bold")
    ax.text(
        x,
        y - 0.035,
        title.replace(" ", "\n", 1),
        ha="center",
        va="center",
        fontsize=7.5,
    )

    if index < len(sequence) - 1:
        ax.annotate(
            "",
            xy=(xs[index + 1] - 0.048, y),
            xytext=(x + 0.048, y),
            arrowprops=dict(arrowstyle="->", linewidth=1.5),
        )

ax.text(0.5, 0.91, "Repository Construction Sequence", ha="center", fontsize=18, fontweight="bold")
ax.text(
    0.5,
    0.17,
    "Each notebook specifies one connected stage of repository development.",
    ha="center",
    fontsize=10.5,
)

savefig("00_repository_construction_sequence.png")
plt.show()
```
